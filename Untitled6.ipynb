```python

import os
os.environ["SDL_VIDEODRIVER"] = "dummy"

from pyvirtualdisplay import Display
display = Display(visible=0, size=(800, 600))
display.start()

import pygame
import sys
import time
from IPython.display import Image, display as ipy_display


N = 250   # Размер сетки по X
M = 60    # Размер сетки по Y
dt = 0.01 #интервал между итерациями
h = 1 #расстояние между ячейками
D = 0.9 #коэф  диффузии
v = 0.5   # скорость конвекции


C = [[0.0 for _ in range(M+2)] for _ in range(N+2)] # ноль ошибок

# Цветовая палитра
palette = [
    (0, 0, 0),       # 0  чернгый
    (0, 0, 128),     # 1  голубой
    (0, 128, 0),     # 2  зеленый
    (0, 128, 128),   # 3  я длтоник
    (128, 0, 0),     # 4  красный
    (128, 0, 128),   # 5  пурпурный
    (128, 128, 0),   # 6  я дальтоник
    (128, 128, 128), # 7  светло-серый
    (192, 192, 192), # 8  темно-серый
    (0, 0, 255),     # 9  светло-голубой
    (0, 255, 0),     # 10 светло-зеленый
    (0, 255, 255),   # 11 я далтоник
    (255, 0, 0),     # 12 светло-карсный
    (255, 0, 255),   # 13 Светло-пурпурный
    (255, 255, 0),   # 14 желтый
    (255, 255, 255)  # 15 белы1
]

def get_color(value: float):

    idx = int(value / 400.0) + 8
    if idx < 0:
        idx = 0
    if idx > 15:
        idx = 15
    return palette[idx]


def Raschet(i, j, t):
    """
    Если (10 < i < 20) и (j > 50) и время t < 5000, добавляем q = 5, иначе q = 0.
    и обнову крутим.
    """
    if (10 < i < 20) and (j > 50) and (t < 5000):
        q = 5.0
    else:
        q = 0.0

    C[i][j] = (
        C[i][j]
        + D * (
            C[i+1][j] - 2 * C[i][j] + C[i-1][j]
            + C[i][j+1] - 2 * C[i][j] + C[i][j-1]
        ) * dt / (h * h)
        + q
        + v * (C[i][j] - C[i+1][j]) / h * dt
    )


pygame.init()
screen_width = 3 * N + 50
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Gazovayz_ataka (Python)")

clock = pygame.time.Clock()

t = 0
iterations = 6000

for iter in range(iterations):
    t += 1


    for i in range(1, N+1):
        for j in range(1, M+1):
            Raschet(i, j, t)


    for j in range(M, 0, -1):
        for i in range(N, 0, -1):
            Raschet(i, j, t)


    for j in range(1, M+1):
        C[N+1][j] = C[N][j]
        C[0][j] = C[1][j]

    # Каждые 100 итераций
    if t % 100 == 0:
        screen.fill((0, 0, 0))

        #пузырьки
        for i in range(1, N+1):
            for j in range(1, M+1):
                color = get_color(C[i][j])
                x_pos = 3 * i
                y_pos = 50 + 3 * j
                pygame.draw.circle(screen, color, (x_pos, y_pos), 2)


        x_plot = int(t / 40)
        indices = [50, 90, 130, 170, 210, 250]
        for idx in indices:
            if 1 <= idx <= N:
                y_plot = 450 - int(2 * C[idx][23])
                if 0 <= x_plot < screen_width and 0 <= y_plot < screen_height:
                    pygame.draw.circle(screen, (255, 255, 255), (x_plot, y_plot), 2)

        pygame.display.flip()


        filename = f"screenshot_{t}.png"
        pygame.image.save(screen, filename)

        ipy_display(Image(filename))
        time.sleep(0.1)

    clock.tick(60)

pygame.quit()
```
